### Removing Outliers

```python
def outliers(data,z_thresold=3):
    outliers = []
    count = 0
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z = (i - mean)/std
        if abs(z) > 3:
            outliers.append(i)
            count+=1
    print('Number of outliers:', count)
    return sorted(outliers)  
```
